```python
listed = drive.ListFile({'q': "title contains '.csv' and 'ID DA PASTA NO GOOGLE DRIVE' in parents"}).GetList()
files = {}
for file in listed:
  files[file['title']] = file['id']
  print('title {}, id {}'.format(file['title'], file['id']))
print(len(files))
```

### CONFIGURAÇÃO DE QUAIS ARQUIVOS DESEJA SUBIR PARA O MySQL

Entre com o nome do arquivo que deseja subir para o MySQL, ou digite '*' para subir todos os arquivos de uma vez

```python
print("Digite o nome do arquivo que deseja subir com a extenção ou '*' para subir todos de uma vez:")
fname = input()

print("Digite 'mysql' ou 'postgresql' para definir em qual banco deseja subir a tabela:")
eng = input()

if eng == 'mysql':
  engine = m_engine
elif eng == 'postgresql':
  engine = p_engine
else:
  print("ERRO: você não digitou a engine do banco corretamente")

if fname == "*":
    subindo = arquivos(engine,files)
    print(subindo)
else:
    fid = files[fname]
    subindo = arquivo(engine,fname,fid)
    print(subindo)
```
